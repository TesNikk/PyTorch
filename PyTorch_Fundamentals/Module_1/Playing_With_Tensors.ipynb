```python
single_distance = torch.tensor(35.0)
print(single_distance.shape)
simple_model = nn.Linear(1,1) # expecs (batch_size,1)
with_batch = single_distance.unsqueeze(0) #added batch dim
print(f"after Squeeing: {with_batch.shape}")
ready_for_model = with_batch.unsqueeze(1) # now ready
print(f"After Unsqueeze: {ready_for_model.shape}")
```

```python
# removing all dimension of size 1
# clean up after batching
squeezed = ready_for_model.squeeze()
print(squeezed.shape)
```

# Indexing and Slicing

```python
predictions = torch.tensor([[14.9],[24.1],[35.6],[45.2]])
first = predictions[0]
print(first)
first_three = predictions[:3]
print(first_three)
#use item() to get a python number
value = predictions[0].item() # works on with tensor with exactly one element
print(value)
```
